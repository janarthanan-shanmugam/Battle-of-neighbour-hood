## Foursquare API Data:

As for four square we should need data about different venues in different neighborhoods of that specific borough. 
so for getting that information we will use "Foursquare" locational information. 
Foursquare is a location data provider with information ( venue names, locations, menus and even photos) about all manner of venues and events within an area of interest. As such, the foursquare location platform will be used as the sole data source since all the stated required information can be obtained through the API. 

Having finded the list of the neigbourhoods then we shoud,connect to the Foursquare API to gather information about venues inside each and every neighborhood. 
each neighbourhoods we should choose the radius to be 100 meter.

The data retrieved from Foursquare contained information of venues within a specified distance of the longitude and latitude of the postcodes. 
The information obtained per venue as follows:

    1. Neighborhood
    2. Neighborhood Latitude
    3. Neighborhood Longitude
    4. Venue
    5. Name of the venue e.g. the name of a store or restaurant
    6. Venue Latitude
    7. Venue Longitude
    8. Venue Category
